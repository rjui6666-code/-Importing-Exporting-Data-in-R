```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Create DataFrame
# -----------------------------
matrix = np.random.randint(1, 101, size=(5, 3))
df = pd.DataFrame(matrix, columns=["A", "B", "C"])
print("DataFrame:\n", df)

# -----------------------------
# 1️⃣ Row-wise operation
# -----------------------------
df["Row_Sum"] = df.sum(axis=1)
df["Row_Mean"] = df.mean(axis=1)
print("\nRow-wise Sum & Mean:\n", df)

# -----------------------------
# 2️⃣ Filtering (condition based)
# -----------------------------
filtered_df = df[df["A"] > 50]
print("\nFiltered (A > 50):\n", filtered_df)

# -----------------------------
# 3️⃣ Sorting
# -----------------------------
sorted_df = df.sort_values(by="B", ascending=False)
print("\nSorted by column B (descending):\n", sorted_df)

# -----------------------------
# 4️⃣ CSV file write & read
# -----------------------------
df.to_csv("my_dataframe.csv", index=False)
print("\nDataFrame saved as my_dataframe.csv")

df_read = pd.read_csv("my_dataframe.csv")
print("\nRead from CSV:\n", df_read)

# -----------------------------
# 5️⃣ Mini visualization
# -----------------------------
plt.figure()
plt.plot(df.index, df["A"])
plt.title("Column A values")
plt.xlabel("Index")
plt.ylabel("A")
plt.show()
```
